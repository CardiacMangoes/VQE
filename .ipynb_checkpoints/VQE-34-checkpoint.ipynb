```python
import numpy as np #basic python numerics package
import pylab #plotting
import copy #convenient(?)
import qiskit
from qiskit import BasicAer #qc simulator
from qiskit.utils import algorithm_globals, QuantumInstance #Tools to run experiment
from qiskit.algorithms import NumPyMinimumEigensolver, VQE #classical values to compare VQE simulation
from qiskit.algorithms.optimizers import SLSQP #classical optimizer to update ansatz (guessed wavefunction)
from qiskit_nature.circuit.library.initial_states import hartree_fock #initial ansatz to optomize
from qiskit_nature.circuit.library import UCCSD #updates initial ansatz with optimization
from qiskit_nature.drivers import PySCFDriver #sets up molecule
from qiskit_nature import Hamiltonian, QubitMappingType #does mapping (for what?)
```

```python
qiskit.__qiskit_version__
```

```python
#initialize a molecule
molecule = 'H .0 .0 -{0}; Li .0 .0 {0}'

#interval of distances we want to calculate energies at
distances = np.arange(0.5, 4.25, 0.25)

#ground state energies calculated by VQE
vqe_energies = []

#calculate initial energies from original HartreeFock guess
hf_energies = []

#Exact energies solved by numpy minimal eigensolver
exact_energies = []
```

```python
#loops over various distances to calculate the VQE at that distance
for i,d in enumerate(distances):
    print('step', i)
    
    #set up experiment
    driver = PySCFDriver(molecule.format(d/2), basis='sto3g')
    qmolecule = driver.run()
    operator = Hamiltonian(qubit_mapping=QubitMappingType.PARITY, two_qubit_reduction=True, 
                           freeze_core=True, orbital_reduction=[-3,-2])
    qubit_op, aux_ops = operator.run(qmolecule)
    
    #exact classical result
    exact_result = NumPyMinimumEigensolver(qubit_op, aux_operators=aux_ops).run()
    exact_result = operator.process_algorithm_result(exact_result)
    
    #VQE
    optimizer = SLSQP(maxiter=1000)
    initial_state = HartreeFock(operator.molecule.info['num_orbitals'], 
                                operator.molecule_info['num_particles'],
                                qubit_mapping = operator._qubit_mapping,
                                two_qubit_reduction = operator._two_qubit_reduction)
    var_form = UCCSD(num_orbitals=operator.molecule_info['num_orbitals'],
                    num_particles=operator.molecule_info['num_particles'],
                    initial_state = initial_state,
                    qubit_mapping=operator._qubit_mapping,
                    two_qubit_reduction=operator._two_qubit_reduction)
    algo = VQE(qubit_op, var_form, optimizer, aux_operators=aux_ops)
    
    vqe_result = algo.run(QuantumInstance(BasicAer.get_backend('statevector_simulator')))
    vqe_result = operator.process_algorithm_result(vqe_result)
    
    exact_energies.append(exact_result.energy)
    vqe_energies.append(vqe_result.energy)
    hf_energies.append(vqe_result.hartree_fock_energy)
```
