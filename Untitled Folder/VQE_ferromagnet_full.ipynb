```python
from qiskit import Aer
from qiskit_nature.converters.second_quantization import QubitConverter
from qiskit_nature.mappers.second_quantization import SpinMapper, DirectMapper, LinearMapper, JordanWignerMapper
from qiskit.opflow import X, Y, Z, I
from qiskit import QuantumCircuit, ClassicalRegister, QuantumRegister
from qiskit.providers.aer import StatevectorSimulator
from qiskit.utils import QuantumInstance
from qiskit_nature.algorithms import VQEUCCFactory

import pylab
import numpy as np
import retworkx as rx
from qiskit_nature.problems.second_quantization.lattice import (
    BoundaryCondition,
    IsingModel,
    FermiHubbardModel,
    HyperCubicLattice,
    Lattice,
    LatticeDrawStyle,
    LineLattice,
    SquareLattice,
    TriangularLattice,
)
```

```python
from qiskit.algorithms import VQE
from qiskit.circuit.library import TwoLocal

tl_circuit = TwoLocal(
    rotation_blocks=["h", "rx"],
    entanglement_blocks="cz",
    entanglement="full",
    reps=2,
    parameter_prefix="y",
)

vqe_solver = VQE(
    ansatz=tl_circuit,
    quantum_instance=QuantumInstance(Aer.get_backend("aer_simulator_statevector")),
)
```

```python
#Locally Coupled
rep = 41
def convertSpin(ham):
    result = None 
    spins = [I, Z, Y, X]
    for c, x, y, z in zip(ham._coeffs, ham.x, ham.y, ham.z):
        spin = x * 3 + y * 2 + z
        pauli = spins[spin[0]]
        for n in range(1, ham.register_length):
            pauli ^= spins[spin[n]]
        pauli *= c
        result = result + pauli if result else pauli
    return result

def makeLatticeIsing(size, cx = 1, cz = 1):
    lattice = LineLattice(size, boundary_condition=BoundaryCondition.PERIODIC)
    ising = IsingModel.uniform_parameters(
        lattice=lattice,
        uniform_interaction=cz, #interaction strength -- "hopping term"
        uniform_onsite_potential=cx
    )
    return convertSpin(ising.second_q_ops())

print(makeLatticeIsing(2, cx = 4).to_matrix())
```

```python
n = 11
results = np.full([rep, n], None)
groundStates = np.full([rep, n], np.inf + 0j)
times = np.full([rep, n], None)

print(groundStates)
```

```python
def findMin(r, size = n):
    #have an array of the coefficients
    for j in range(rep):
        cx = round(-2.0 + (4 * j/(rep-1)), 5)
        print(cx)
        for i in range(1, size):
            for _ in range(r):
                ising = makeLatticeIsing(i, cx = cx)
                if ising:
                    result = vqe_solver.compute_minimum_eigenvalue(ising)
                    if np.absolute(result.eigenvalue) < np.absolute(groundStates[j][i]):
                        results[j][i] = result
                        groundStates[j][i] = result.eigenvalue
                        times[j][i] = result.optimizer_time
                        print(f"Minimum {i:>4}: {np.round(result.eigenvalue, 3):<8} time: {round(result.optimizer_time, 2)}")
                    else:
                        print(f"Processed {i:>2},          time: {round(result.optimizer_time, 2)}")

findMin(10)
```
